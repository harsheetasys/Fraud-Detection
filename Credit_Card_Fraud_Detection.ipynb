```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from torch_geometric.data import Data
from torch_geometric.nn import GCNConv
from sklearn.neighbors import kneighbors_graph
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score,accuracy_score
import seaborn as sns

def create_fraud_dataset(num_samples=1000):
    np.random.seed(42)
    data = {
        'Time': np.arange(num_samples),
        'V1': np.random.normal(0, 1, num_samples),
        'V2': np.random.normal(0, 1, num_samples),
        'V3': np.random.normal(0, 1, num_samples),
        'V4': np.random.normal(0, 1, num_samples),
        'V5': np.random.normal(0, 1, num_samples),
        'V6': np.random.normal(0, 1, num_samples),
        'V7': np.random.normal(0, 1, num_samples),
        'V8': np.random.normal(0, 1, num_samples),
        'V9': np.random.normal(0, 1, num_samples),
        'V10': np.random.normal(0, 1, num_samples),
        'V11': np.random.normal(0, 1, num_samples),
        'V12': np.random.normal(0, 1, num_samples),
        'V13': np.random.normal(0, 1, num_samples),
        'V14': np.random.normal(0, 1, num_samples),
        'V15': np.random.normal(0, 1, num_samples),
        'V16': np.random.normal(0, 1, num_samples),
        'V17': np.random.normal(0, 1, num_samples),
        'V18': np.random.normal(0, 1, num_samples),
        'V19': np.random.normal(0, 1, num_samples),
        'V20': np.random.normal(0, 1, num_samples),
        'V21': np.random.normal(0, 1, num_samples),
        'V22': np.random.normal(0, 1, num_samples),
        'V23': np.random.normal(0, 1, num_samples),
        'V24': np.random.normal(0, 1, num_samples),
        'V25': np.random.normal(0, 1, num_samples),
        'V26': np.random.normal(0, 1, num_samples),
        'V27': np.random.normal(0, 1, num_samples),
        'V28': np.random.normal(0, 1, num_samples),
        'Amount': np.random.uniform(0, 500, num_samples),
        'Class': np.random.choice([0, 1], size=num_samples, p=[0.99, 0.01])
    }
    return pd.DataFrame(data)

# Create dataset
df = create_fraud_dataset(500)  # 500 transactions for quick training
print(f"Fraud rate: {df['Class'].mean():.2%}")

# Step 4: Build transaction graph
features = df.drop(['Class'], axis=1).values
labels = df['Class'].values

# Create graph edges (connect similar transactions)
edge_matrix = kneighbors_graph(features[:, 1:4],  # Use V1-V3 for similarity
                               n_neighbors=5,
                               mode='connectivity',
                               include_self=False)
edge_index = edge_matrix.nonzero()
edge_index = torch.tensor(np.array(edge_index), dtype=torch.long)

# Create PyG Data object
data = Data(
    x=torch.tensor(features, dtype=torch.float),
    edge_index=edge_index,
    y=torch.tensor(labels, dtype=torch.long)
)

# Step 5: Define GNN model
class FraudGNN(nn.Module):
    def __init__(self, num_features, hidden_channels, num_classes):
        super().__init__()
        self.conv1 = GCNConv(num_features, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, num_classes)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=0.5, training=self.training)
        x = self.conv2(x, edge_index)
        return F.log_softmax(x, dim=1)

# Step 6: Train model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = FraudGNN(
    num_features=features.shape[1],
    hidden_channels=32,
    num_classes=2
).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)
criterion = nn.NLLLoss()

# Move data to device
data = data.to(device)

# Training loop
model.train()
for epoch in range(20):
    optimizer.zero_grad()
    out = model(data.x, data.edge_index)
    loss = criterion(out, data.y)
    loss.backward()
    optimizer.step()
    print(f'Epoch {epoch+1:02d}, Loss: {loss.item():.4f}')

# Step 7: Evaluate model
model.eval()
with torch.no_grad():
    pred = model(data.x, data.edge_index).argmax(dim=1)
    y_true = data.y.cpu().numpy()
    y_pred = pred.cpu().numpy()

# Metrics
print("\nClassification Report:")
print(classification_report(y_true, y_pred))

print(f"ROC AUC: {roc_auc_score(y_true, y_pred):.4f}")
print(f"Accuracy: {accuracy_score(y_true, y_pred):.4f}")
# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Step 8: Generate synthetic fraud data (GAN-like approach)
def generate_synthetic_fraud(model, num_samples=20):
    """Generate synthetic fraud samples using model-based perturbation"""
    model.eval()
    fraud_indices = np.where(y_true == 1)[0]
    synthetic_samples = []

    for _ in range(num_samples):
        # Select a random fraud sample
        idx = np.random.choice(fraud_indices)
        sample = data.x[idx].clone().detach()

        # Add noise to create new sample
        noise = torch.randn_like(sample) * 0.1
        synthetic_sample = sample + noise

        synthetic_samples.append(synthetic_sample.cpu().numpy())

    return np.array(synthetic_samples)

# Generate synthetic fraud cases
synthetic_fraud = generate_synthetic_fraud(model)
print(f"Generated {len(synthetic_fraud)} synthetic fraud samples")

# Step 9: Visualize results
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(df['Amount'][df['Class'] == 0], bins=30, alpha=0.7, label='Legitimate')
plt.hist(df['Amount'][df['Class'] == 1], bins=30, alpha=0.7, label='Fraud')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.legend()
plt.title('Transaction Amount Distribution')

plt.subplot(1, 2, 2)
fraud_scores = model(data.x, data.edge_index)[:, 1].detach().cpu().numpy()
plt.hist(fraud_scores[y_true == 0], bins=30, alpha=0.7, label='Legitimate')
plt.hist(fraud_scores[y_true == 1], bins=30, alpha=0.7, label='Fraud')
plt.xlabel('Fraud Score')
plt.ylabel('Frequency')
plt.legend()
plt.title('Model Fraud Scores')
plt.tight_layout()
plt.show()
```
